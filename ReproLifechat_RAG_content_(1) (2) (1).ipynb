```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import MessagesPlaceholder

prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            '''
You are a sex life expert who provides guidance and answers to individuals about their reproductive system, sexual health, and relationships. Use the following approach for handling
different types of queries:

Sex Life and Reproductive Health Queries:

For all sex life-related questions, use your pre-trained knowledge and the RAG content to provide accurate and thoughtful answers.
 for Weather-Related Queries:

If a query is related to weather or time, use the weather_tool to retrieve weather information for a specific location. The input should include both the city and country.
If either the city or country is missing in the prompt, inform the user that they must provide both.
If the provided city or country does not exist or is incorrect, prompt the user to provide valid information.
for Unrelated Queries (Football, Music, etc.):

For any queries unrelated to sex life or weather (such as sports, music, etc.), use external tools like the tavily_search tool to fetch links to relevant information.
Make sure to guide the user to where they can find appropriate resources for these topics.

      '''
        ),
        MessagesPlaceholder(variable_name="chat_history"),
        ("user", "{input}"),
        MessagesPlaceholder(variable_name="agent_scratchpad"),
    ]
)
```

```python
from langchain.agents.output_parsers.openai_tools import OpenAIToolsAgentOutputParser
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentExecutor
from langchain.agents.format_scratchpad.openai_tools import format_to_openai_tool_messages

agent = (
    {
        "input": lambda x: x["input"],
        "agent_scratchpad": lambda x: format_to_openai_tool_messages(
            x["intermediate_steps"]
        ),
        "chat_history": lambda x: x["chat_history"]
    }
    | prompt
    | llm_with_tools
    | OpenAIToolsAgentOutputParser()
)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
from langchain_core.messages import AIMessage, HumanMessage

chat_history = []
```

```python

while True:
    print("")
    print("")
    print("")
    print("____________________________________________________________________________________________________________________________" *2)
    input1 = input("Enter your prompt (type 'exit' to quit): ")

    if input1.lower() == "exit":
        print("Exiting the chat...")
        break

    result = agent_executor.invoke({"input": input1, "chat_history": chat_history, "tool": "llm_tool"})

    chat_history.extend(
        [
            HumanMessage(content=input1),
            AIMessage(content=result["output"]),
        ]
    )

    print("AI response:", result["output"])
```
